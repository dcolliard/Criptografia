```python
p = 17
alfa = 3
```

```python
x = 13

x_upper = (alfa**x) % p

print("X = (alfa**x) % p")
print("X = ("+str(alfa)+"**"+str(x)+") % "+str(p))
print("X = "+str((alfa**x))+" % "+str(p))

print("X = " + str(x_upper))
```

```python
y = 11

y_upper = (alfa**y) % p

print("Y = (alfa**y) % p")
print("Y = ("+str(alfa)+"**"+str(y)+") % "+str(p))
print("Y = "+str((alfa**y))+" % "+str(p))

print("Y = " + str(y_upper))
```

```python
k = (y_upper**x) % p

print("k = (Y**x) % p")
print("k = ("+str(y_upper)+"**"+str(x)+") % "+str(p))
print("k = "+str(y_upper**x)+" % "+str(p))

print("k = " + str(k))
```

```python
k = (x_upper**y) % p

print("k = (X**y) % p")
print("k = ("+str(x_upper)+"**"+str(y)+") % "+str(p))
print("k = "+str(x_upper**y)+" % "+str(p))

print("k = " + str(k))
```
