```python
p = 7
q = 11

print("p = " + str(p))
print("q = " + str(q))
```

```python
n = p * q
print("n = " + str(n))

phi = (p-1) * (q-1)
print("𝜑(𝑛) = " + str(phi))
```

```python
e = 13
print("e = " + str(e))
```

```python
d = 37
print("d = " + str(d))
print("")
print("𝑒𝑑 ≡ 1 (𝑚𝑜𝑑 𝜑(𝑛))")
print("("+str(e)+")("+str(d)+") ≡ 1 (𝑚𝑜𝑑 "+str(phi)+")")
print(str(e*d)+" ≡ 1 (𝑚𝑜𝑑 "+str(phi)+")")
print(str((e*d)%phi)+" ≡ 1 (𝑚𝑜𝑑 "+str(phi)+")")
```

```python
c = (m ** e) % n

print("c ≡ 𝑚**𝑒 (𝑚𝑜𝑑 𝑛)")
print("c ≡ "+str(m)+"**"+str(e)+" (𝑚𝑜𝑑 "+str(n)+")")
print("c ≡ "+str(m**e)+" (𝑚𝑜𝑑 "+str(n)+")")
print("c ≡ "+str((m**e)%n)+" (𝑚𝑜𝑑 "+str(n)+")")
```

```python
m = (c ** d) % n

print("m ≡ c**d (𝑚𝑜𝑑 𝑛)")
print("m ≡ "+str(c)+"**"+str(d)+" (𝑚𝑜𝑑 "+str(n)+")")
print("m ≡ "+str(c**d)+" (𝑚𝑜𝑑 "+str(n)+")")
print("m ≡ "+str((c**d)%n)+" (𝑚𝑜𝑑 "+str(n)+")")
print("m = "+str(m))
```
